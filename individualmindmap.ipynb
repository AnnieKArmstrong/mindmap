# Import Libraries

```python
import pandas as pd
import openpyxl
import xlrd
import matplotlib.pyplot as plt
import numpy as np
import networkx
pd.options.display.max_rows = 400
plt.rcParams["figure.figsize"] = [10, 6]
```

## Read Data

```python
X = pd.ExcelFile("C:/Users/Annie Armstrong/Documents/Github/mindmap/grp_edgelists/grp-edgelist.xlsx")
mindmap_df = X.parse("Individual lists")
grpmindmap_df = X.parse("Group list")

#Group DataFrames
grpseptmw_df = grpmindmap_df.query('Group == "MW" & Month == "September"')
grpdecmw_df = grpmindmap_df.query('Group == "MW" & Month == "December"')
grpsepttth_df = grpmindmap_df.query('Group == "TTh" & Month == "September"')
grpdectth_df = grpmindmap_df.query('Group == "TTh" & Month == "December"')

#Individual Map dataframes
septmw_df = mindmap_df.query('Group == "MW" & Month == "September"')
decmw_df = mindmap_df.query('Group == "MW" & Month == "December"')
septtth_df = mindmap_df.query('Group == "TTh" & Month == "December"')
dectth_df = mindmap_df.query('Group == "TTh" & Month == "December"')
```

## Create Network maps

```python
G = networkx.from_pandas_edgelist(decmw_df, 'Edge', 'Target')
networkx.write_graphml(G, 'decmw.graphml')
plt.figure(figsize=(8,8))
networkx.draw(G, with_labels=True, node_color='skyblue', width=.3, font_size=8)
print('decmw', G.number_of_nodes(), G.number_of_edges())
networkx.degree(G)

G = networkx.from_pandas_edgelist(septtth_df, 'Edge', 'Target')
networkx.write_graphml(G, 'septtth.graphml')
plt.figure(figsize=(8,8))
networkx.draw(G, with_labels=True, node_color='skyblue', width=.3, font_size=8)
print('septtth', G.number_of_nodes(), G.number_of_edges())
networkx.degree(G)

G = networkx.from_pandas_edgelist(dectth_df, 'Edge', 'Target')
networkx.write_graphml(G, 'dectth.graphml')
plt.figure(figsize=(8,8))
networkx.draw(G, with_labels=True, node_color='skyblue', width=.3, font_size=8)
print('dectth', G.number_of_nodes(), G.number_of_edges())
networkx.degree(G)
```

# September Monday Wednesday Maps

```python
#September Monday Wednesday Individual

septmwG = networkx.from_pandas_edgelist(septmw_df, 'Edge', 'Target')
networkx.write_graphml(septmwG, 'septmw.graphml')
plt.figure(figsize=(8,8))
networkx.draw(septmwG, with_labels=True, node_color='skyblue', width=.3, font_size=8)
print('septmw', septmwG.number_of_nodes(), septmwG.number_of_edges())
networkx.degree(septmwG)

degrees = dict(networkx.degree(septmwG))
networkx.set_node_attributes(septmwG, name='degree', values=degrees)
degree_df = pd.DataFrame(septmwG.nodes(data='degree'), columns=['node', 'degree'])
degree_df = degree_df.sort_values(by='degree', ascending=False)
degree_df.head(10)
```

```python
num_nodes_to_inspect = 10
degree_df[:num_nodes_to_inspect].plot(x='node', y='degree', kind='barh').invert_yaxis()
```

```python
#September Group Map Information

G = networkx.from_pandas_edgelist(groupseptmw_df, 'Edge', 'Target')
networkx.write_graphml(G, 'groupseptmwG.graphml')
print('Groupseptmw', G.number_of_nodes(), G.number_of_edges())
plt.figure(figsize=(8,8))
networkx.draw(G, with_labels=True, node_color= 'skyblue', width = .3, font_size=8)

degrees = dict(networkx.degree(G))
networkx.set_node_attributes(G, name='degree', values=degrees)
degree_df = pd.DataFrame(G.nodes(data='degree'), columns=['node', 'degree'])
degree_df = degree_df.sort_values(by='degree', ascending=False)
degree_df.head(10)

num_nodes_to_inspect = 10
degree_df[:num_nodes_to_inspect].plot(x='node', y='degree', kind='barh').invert_yaxis()
```
